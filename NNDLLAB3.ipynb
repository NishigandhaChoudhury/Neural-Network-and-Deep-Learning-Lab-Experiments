```python
vector1 = [10, 20, 30, 40]
vector2 = [2, 4, 5, 8]
```

```python
addition = [a + b for a, b in zip(vector1, vector2)]
```

```python
subtraction = [a - b for a, b in zip(vector1, vector2)]
```

```python
multiplication = [a * b for a, b in zip(vector1, vector2)]
```

```python
division = [a / b for a, b in zip(vector1, vector2)]
```

```python
print("Vector 1:", vector1)
print("Vector 2:", vector2)
print("Addition:", addition)
print("Subtraction:", subtraction)
print("Multiplication:", multiplication)
print("Division:", division)
```
