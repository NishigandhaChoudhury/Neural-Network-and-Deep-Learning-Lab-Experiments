```python
import numpy as np
```

```python
tensor_0d = np.array(5)
print(tensor_0d)
print("0-D Tensor (Scalar):")
print("Dimensions:", tensor_0d.ndim)
print("Shape:", tensor_0d.shape)
print()
```

```python
tensor_1d = np.array([1, 2, 3, 4])
print("1-D Tensor (Vector):")
print(tensor_1d)
print("Dimensions:", tensor_1d.ndim)
print("Shape:", tensor_1d.shape)
print()
```

```python
tensor_2d = np.array([[1, 2, 3],
                      [4, 5, 6]])
print("2-D Tensor (Matrix):")
print(tensor_2d)
print("Dimensions:", tensor_2d.ndim)
print("Shape:", tensor_2d.shape)
print()
```

```python
tensor_3d = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])
print("3-D Tensor:")
print(tensor_3d)
print("Dimensions:", tensor_3d.ndim)
print("Shape:", tensor_3d.shape)
print()
```

```python
tensor_4d = np.zeros((2, 3, 4, 5))
print("4-D Tensor:")
print("Dimensions:", tensor_4d.ndim)
print("Shape:", tensor_4d.shape)
```
